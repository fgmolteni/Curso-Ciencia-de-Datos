
Con la funcion `read_csv()` los argumento que podemos agregar:
* `sep` (que delimitador utilazara es set de datos).
* `dtype` (establese que columna tendra que tipo de datos).
* `header` se especifica que la caberecea de las filas sera la que designemos.
* Con `name` designaremos los nombres de las columnas.
* `kiprows` es el valor entero que se saltan a la hora de leer el fichero.
* `index_col=numero` indicara que las columnas designada sera un fila.
* `skip_blank_lines=True` las lineas en blanco se saltaran en vez de ser valores **NaN**
* `na_filter=True` eliminaria toda aquellas **filas** que tendrian valores **desconocidos**(**NaN** o string vacios).

```python
pd.read_csv(
            filepath_or_buffer=r'repositorio\datasets\titanic\titanic3.csv',
            sep=',',
            header=0,
            dtype={'age': np.float64,},
            skiprows=12,
            index_col=None,
            skip_blank_lines=True,
            na_filter=False
            )
```
